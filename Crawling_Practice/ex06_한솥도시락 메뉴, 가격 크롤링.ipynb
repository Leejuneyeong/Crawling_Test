### 예외처리 하는방법
- try문에서 코드를 실행하다가 에러가 나면 except 를 실행

```python
# try except를 사용하는 이유!
# 코드 실행시 에러가 발생하면 밑에 작성된 코드들은 중간에 멈춤!
# 멈춤을 막고 계속해서 아래 코드를 실행할 수 있도록 예외처리를 진행
try :
    # 에러가 나기 전까지 실행할 부분을 try 부분에 작성
    for i in range(30):
        btn_more = driver.find_element(By.CLASS_NAME, 'c_05')
        btn_more.click()
except : # 에러가 나면 출력해주는 것을 except 부분에 작성
    print('클릭완료')
```

```python
# 한솥 사이트 크롤링 해오기 
driver = wb.Chrome()
# 드라이버를 홈페이지에 연결
driver.get('https://www.hsd.co.kr/menu/menu_list')

try :
    for i in range(30):
        btn_more = driver.find_element(By.CLASS_NAME, 'c_05')
        btn_more.click()
        time.sleep(2) # (초)
except : 
    print('클릭완료')
#이 아래부터는 서버와의 연결이 끊긴 이후 -> time.sleep()줄 필요없다
#이미 받아온 정보이기 때문이다.
# 컴퓨터가 알아볼 수 있도록 BeautifulSoup 해줌
soup = bs(driver.page_source,'lxml')
menu = soup.select('h4.h.fz_03')
price = soup.select('div.item-price>strong')
```

```python
#데이터 활용
#순수한 글자정보만 담긴 리스트 제작
#데이터프레임제작
#데이터프레임이 한글이 깨지지 않게 csv 파일로 저장
```

```python
import pandas as pd
num_list = [] 
menu_list = []
price_list=[]
for i in range(len(menu)) :
    num_list.append(i+1)
    menu_list.append(menu[i].text)
    price_list.append(price[i].text)
# dic 이 편함    
dic = {'순번':num_list, '메뉴': menu_list, '가격':price_list}
total = pd.DataFrame(dic)
total.set_index('순번',inplace=True)
total
```

```python
#데이터프레임을 csv 파일로 바꿔줌
total.to_csv('한솥메뉴표.csv',encoding='euc-kr')
```
